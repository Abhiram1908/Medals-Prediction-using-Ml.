# Accuracy Metrics

We will use mean squared error.This is a good default regression accuracy metric.Its the average of squared differences between the actual results and the predictions we made.

Model used is the linear regression to predict the data using linear hypothesis through data in various steps .

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
predictors = ["athletes","prev_medals"]
target = "medals"
```

```python
reg.fit(train[predictors],train["medals"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions
```

```python
test["predictions"] = predictions
```

```python
test
```

```python
test.loc[test["predictions"] < 0, "predictions"] = 0
```

```python
test["predictions"] = test["predictions"].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test["medals"], test["predictions"])
```

```python
error
```
